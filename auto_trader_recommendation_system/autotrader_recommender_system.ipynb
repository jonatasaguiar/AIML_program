### 2. Exploratory Data Analysis and Data Visualization

```python
import pandas as pd 

ratings = pd.read_csv('files/ratings_full.csv')
car_years = list(ratings['car_year'].unique())
```

```python
ratings.info()
```

```python
# Checks if there are null values in 'df' dataset
ratings.isnull().sum()
```

```python
# Setting parameters
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("white")
plt.figure(figsize=(8,5), dpi= 80)
%matplotlib inline
```

```python
# creating filter for years
from tqdm import tqdm 
car_grouped_list = []
top_reviewed_list = []
top_rated_list = []
top_user_list = []
for year in tqdm(car_years):
    # Creates a dataframe that contains the mean of 
    # the total rating for each car rating
    path_read = 'files/ratings_full_{}.csv'.format(year)
    df = pd.read_csv(path_read)
    rating = pd.DataFrame(df.groupby('car')['rating'].mean())
    rating.rename(columns={'rating':'rating_mean'}, inplace=True) # changing the 'rating' column name for 'rating_mean'

    # Adds a 'count' column to the rating dataframe 
    # that displays the counts for the number of reviews
    rating['rating_count'] = pd.DataFrame(df.groupby('car')['rating'].count())
    
    car_grouped_list.append(rating)
    
    # top reviewed cars
    top_reviewed = rating.sort_values('rating_count', ascending=False)
    top_reviewed_list.append(top_reviewed)
    
    # top rated cars
    top_rated = rating.sort_values('rating_mean', ascending=False)
    top_rated_list.append(top_rated)
    
    # Groups by user 
    user_grouped = pd.DataFrame(df.groupby(by='userId')['car'].count())
    user_grouped.rename(columns={'car':'review_count'}, inplace=True)
    user_grouped = user_grouped.sort_values('review_count', ascending=False)
    top_user_list.append(user_grouped)
```

```python
# Prints the distribution and adds a smooth density curve as 
# a line for the number of observations of the count of
# ratings mean
def distribution(year):
    index = car_years.index(year)
    df_temporary = car_grouped_list[index]
    sns.histplot(df_temporary['rating_mean'], bins=15, color="dodgerblue", kde=True, label="ratings mean for {}".format(year))
    minimum = df_temporary['rating_mean'].min()
    maximum = df_temporary['rating_mean'].max()
    plt.xlim(minimum, maximum)
    plt.legend()
    plt.show()
```

```python
for year in car_years:
    distribution(year)
```

```python
# Sort the most popular cars in the year
def top_reviewed(year):
    index = car_years.index(year)
    df_temporary = top_reviewed_list[index]
    return df_temporary
```

```python
def user_name(user_id, year):
    path_read = 'files/ratings_full_{}.csv'.format(year)
    df_temporary = pd.read_csv(path_read)
    df_user_name = df_temporary[df_temporary['userId']==user_id]
    name = df_user_name['user'].values[0]
    return name
```

### Item-based (item-item) collaborative filtering

Looking up 10 similar cars to Dodge Ram Pickup(2009) using 'person_baseline' similarity

```python
import pandas as pd
from surprise import KNNBasic
import heapq
from collections import defaultdict
from operator import itemgetter
from surprise.model_selection import LeaveOneOut
```

```python
from surprise import KNNBaseline, Dataset, Reader
```

```python
# importing required modules from Surprise
from surprise import KNNBaseline, Dataset, Reader

# inputs
car = 'Dodge Ram Pickup'
k = 10
year = 2012

similarity = ['pearson_baseline', 'cosine']
similarity_type = similarity[0]

path = 'files/rid_name_car_{}.csv'.format(year)

rid_name = pd.read_csv(path, dtype = str)
rid = (rid_name['rid'].values.tolist())
name = (rid_name['name'].values.tolist())
rid_to_name = dict(zip(rid, name))
name_to_rid = dict(zip(name, rid))

ratingsPath = 'files/ratings_{}.csv'.format(year)
reader = Reader(line_format='user item rating', sep=',', skip_lines=1)
data = Dataset.load_from_file(ratingsPath, reader=reader)

trainset = data.build_full_trainset()
sim_options = {'name': similarity_type, 
               'user_based': False
              }
algo = KNNBaseline(sim_options=sim_options)
algo.fit(trainset)

# Read the mappings raw id <-> car name
rid_to_name = rid_to_name
name_to_rid = name_to_rid

# Retrieve inner id of the car BMW 3 Series
car_raw_id = name_to_rid[car]
car_inner_id = algo.trainset.to_inner_iid(car_raw_id)

# Retrieve inner ids of the nearest neighbors of BMW 3 Series.
car_neighbors = algo.get_neighbors(car_inner_id, k=k)

# Convert inner ids of the neighbors into names.
car_neighbors = (algo.trainset.to_raw_iid(inner_id) for inner_id in car_neighbors)
car_neighbors = (rid_to_name[rid] for rid in car_neighbors)

print()
print('The {} nearest neighbors of {}({}) from the same year are:'.format(k, car, year))
for car in car_neighbors:
    print(car)
```
