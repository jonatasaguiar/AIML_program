# Saving images

```python
def get_full_link(link, image_number):    
    complement = '?defaultImage=image_na_EN&fmt=jpg&fit=constrain,1&wid=700&hei=700'
    link = link[:-1]
    full_link = "{}{}{}".format(link, image_number, complement)
    return full_link
```
